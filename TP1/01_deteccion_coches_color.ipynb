```python
%matplotlib qt

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Cargar la imagen color
img_color = cv.imread('coches_color.jpg')
imgRGB = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)   

#plt.figure("Imagen de coches")
#plt.imshow(imgRGB)
#plt.show()

# Convertir la imagen de BGR a HSV
imgHSV = cv.cvtColor(img_color, cv.COLOR_BGR2HSV)
#plt.figure("Imagen en espacio de color HSV")
#plt.imshow(imgHSV)
#plt.show()

# Hacemos una descripcion estadistica para coche naranja
muestra = imgHSV[184:196, 152:162,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche naranja: {color_h_medio}')
SPAN_RANGE = 1
color_l = (color_h_medio - SPAN_RANGE, 50, 50)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_naranja = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche naranja segmentado")
plt.imshow(img_segmentada_coche_naranja)
plt.show()

# Hacemos una descripcion estadistica para coche Verde
muestra = imgHSV[98:110, 222:244,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche verde: {color_h_medio}')
SPAN_RANGE = 5
color_l = (color_h_medio - SPAN_RANGE, 50, 50)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_verde = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche verde segmentado")
plt.imshow(img_segmentada_coche_verde)
plt.show()

# Hacemos una descripcion estadistica para coche turqueza
muestra = imgHSV[138:148, 355:389,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche turqueza: {color_h_medio}')
SPAN_RANGE = 1
color_l = (color_h_medio - SPAN_RANGE, 0, 0)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_turqueza = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche turqueza segmentado")
plt.imshow(img_segmentada_coche_turqueza)
plt.show()


# Hacemos una descripcion estadistica para coche blanco
muestra = imgHSV[82:91, 43:82,:]
color_h_medio, s, v = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche blanco: {color_h_medio}, s={s}, v={v} ')
color_l = (0, 0, 200)
color_u = (179, 50, 255)

mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_blanco = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche blanco segmentado")
plt.imshow(img_segmentada_coche_blanco)
plt.show()


# Hacemos una descripcion estadistica para coche azul
muestra = imgHSV[95:97, 130:160,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche azul: {color_h_medio}')
SPAN_RANGE = 1
color_l = (color_h_medio - SPAN_RANGE, 80, 80)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_azul = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche azul segmentado")
plt.imshow(img_segmentada_coche_azul)
plt.show()

# Hacemos una descripcion estadistica para coche Rojo
muestra = imgHSV[75:88, 198:235,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche rojo: {color_h_medio}')
SPAN_RANGE = 100
color_l = (color_h_medio - SPAN_RANGE, 100, 100)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
color_l = (0, 100, 100)
color_u = (10, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_rojo = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche rojo segmentado")
plt.imshow(img_segmentada_coche_rojo)
plt.show()

# Hacemos una descripcion estadistica para coche Amarillo
muestra = imgHSV[77:86, 278:300,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche amarillo: {color_h_medio}')
SPAN_RANGE = 1
color_l = (color_h_medio - SPAN_RANGE, 50, 50)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_amarillo = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche amarillo segmentado")
plt.imshow(img_segmentada_coche_amarillo)
plt.show()


# Hacemos una descripcion estadistica para coche Marron
muestra = imgHSV[82:91, 346:380,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche marron: {color_h_medio}')
SPAN_RANGE = 1
color_l = (color_h_medio - SPAN_RANGE, 50, 50)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_marron = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche marron segmentado")
plt.imshow(img_segmentada_coche_marron)
plt.show()

# Hacemos una descripcion estadistica para coche Verde oscuro
muestra = imgHSV[97:103, 412:426,:]
color_h_medio, _, _ = np.mean(muestra, axis=(0,1))
print(f'Valor medio de la componente Hue coche verde oscuro: {color_h_medio}')
SPAN_RANGE = 1
color_l = (color_h_medio - SPAN_RANGE, 50, 50)
color_u = (color_h_medio + SPAN_RANGE, 255, 255)
mask = cv.inRange(src=imgHSV, lowerb=color_l, upperb=color_u)
#muestra = cv.cvtColor(muestra, cv.COLOR_HSV2BGR)
img_segmentada_coche_verdeoscuro = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("coche verde oscuro segmentado")
plt.imshow(img_segmentada_coche_verdeoscuro)
plt.show()



img_segmentada_suma = img_segmentada_coche_naranja + img_segmentada_coche_verde + img_segmentada_coche_turqueza + img_segmentada_coche_blanco + img_segmentada_coche_azul + img_segmentada_coche_rojo + img_segmentada_coche_amarillo + img_segmentada_coche_marron + img_segmentada_coche_verdeoscuro


plt.figure("Suma total de la segmentacion de coches")
plt.imshow(img_segmentada_suma)
plt.title('coches segmentado total')
plt.show()
```

```python
# Calculamos un modelo estadístico para caracterizar del coche naranja
img_color = cv.imread('coches_color.jpg')
imgRGB = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)   

plt.figure("Imagen de coches")
plt.imshow(imgRGB)
plt.show()
# muestra del coche naranja en el espacio RGB
img_muestra = imgRGB[184:196, 152:162,:]
color_mean_n, color_std_n = cv.meanStdDev(img_muestra)
print(f'Media rgb_Naranja: {color_mean_n.flatten().tolist()}')
print(f'Desvio rgb_Naranja: {color_std_n.flatten().tolist()}')
plt.figure("Muestra coche naranja")
plt.imshow(img_muestra)
plt.show()



# Tomamos una muestra del coche Verde
img_muestra = imgRGB[98:110, 222:244,:]
color_mean_v, color_std_v = cv.meanStdDev(img_muestra)
print(f'Media rgb_Verde: {color_mean_v.flatten().tolist()}')
print(f'Desvio rgb_Verde: {color_std_v.flatten().tolist()}')
plt.figure("Muestra Verde")
plt.imshow(img_muestra)
plt.show()



# Tomamos una muestra del coche turqueza
img_muestra = imgRGB[138:148, 355:389,:]
color_mean_tur, color_std_tur = cv.meanStdDev(img_muestra)
print(f'Media rgb_Turquesa: {color_mean_tur.flatten().tolist()}')
print(f'Desvio rgb_Turquesa: {color_std_tur.flatten().tolist()}')
plt.figure("Muestra Turqueza")
plt.imshow(img_muestra)
plt.show()

# Tomamos una muestra del coche blanco
img_muestra = imgRGB[82:91, 43:82,:]
color_mean_b, color_std_b = cv.meanStdDev(img_muestra)
print(f'Media rgb_Blanco: {color_mean_b.flatten().tolist()}')
print(f'Desvio rgb_Blanco: {color_std_b.flatten().tolist()}')
plt.figure("Muestra Blanco")
plt.imshow(img_muestra)
plt.show()


# Tomamos una muestra del coche Azul
img_muestra = imgRGB[95:97, 130:160,:]
color_mean_a, color_std_a = cv.meanStdDev(img_muestra)
print(f'Media rgb_Azul: {color_mean_a.flatten().tolist()}')
print(f'Desvio rgb_Azul: {color_std_a.flatten().tolist()}')
plt.figure("Muestra Azul")
plt.imshow(img_muestra)
plt.show()



# Tomamos una muestra del coche Rojo
img_muestra = imgRGB[75:88, 198:235,:]
color_mean_r, color_std_r = cv.meanStdDev(img_muestra)
print(f'Media rgb_Rojo: {color_mean_r.flatten().tolist()}')
print(f'Desvio rgb_Rojo: {color_std_r.flatten().tolist()}')
plt.figure("Muestra Rojo")
plt.imshow(img_muestra)
plt.show()


# Tomamos una muestra del coche Amarillo
img_muestra = imgRGB[77:86, 278:300,:]
color_mean_am, color_std_am = cv.meanStdDev(img_muestra)
print(f'Media rgb_Amarillo: {color_mean_am.flatten().tolist()}')
print(f'Desvio rgb_Amarillo: {color_std_am.flatten().tolist()}')
plt.figure("Muestra Amarillo")
plt.imshow(img_muestra)
plt.show()


# Tomamos una muestra del coche Marron
img_muestra = imgRGB[82:91, 346:380,:]
color_mean_m, color_std_m = cv.meanStdDev(img_muestra)
print(f'Media rgb_Marron: {color_mean_m.flatten().tolist()}')
print(f'Desvio rgb_Marron: {color_std_m.flatten().tolist()}')
plt.figure("Muestra Marron")
plt.imshow(img_muestra)
plt.show()

# Tomamos una muestra del coche Verde Oscuro
img_muestra = imgRGB[97:103, 412:426,:]
color_mean_vo, color_std_vo = cv.meanStdDev(img_muestra)
print(f'Media rgb_verde oscuro: {color_mean_vo.flatten().tolist()}')
print(f'Desvio rgb_Verde oscuro: {color_std_vo.flatten().tolist()}')
plt.figure("Muestra Verde Oscuro")
plt.imshow(img_muestra)
plt.show()

```

```python
# Identificamos los puntos que estén cerca de la media en un radio menor 
# que 3 sigmas 

#coche Naranja n
n_sigma = 3
mask = cv.inRange(
    imgRGB, color_mean_n-color_std_n * n_sigma,  color_mean_n + color_std_n * n_sigma)
img_segmentada_coche_naranja = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche naranja")
plt.imshow(img_segmentada_coche_naranja)
plt.show()

#coche Verde v
mask = cv.inRange(
    imgRGB, color_mean_v-color_std_v * n_sigma,  color_mean_v + color_std_v * n_sigma)
img_segmentada_coche_verde = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche Verde")
plt.imshow(img_segmentada_coche_verde)
plt.show()

#coche Turqueza tur
mask = cv.inRange(
    imgRGB, color_mean_tur-color_std_tur * n_sigma,  color_mean_tur + color_std_tur * n_sigma)
img_segmentada_coche_turqueza = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche turqueza")
plt.imshow(img_segmentada_coche_turqueza)
plt.show()

#coche Blanco b
mask = cv.inRange(
    imgRGB, color_mean_b-color_std_b * n_sigma,  color_mean_b + color_std_b * n_sigma)
img_segmentada_coche_blanco = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche blanco")
plt.imshow(img_segmentada_coche_blanco)
plt.show()

#coche Azul a
mask = cv.inRange(
    imgRGB, color_mean_a-color_std_a * n_sigma,  color_mean_a + color_std_a * n_sigma)
img_segmentada_coche_azul = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche azul")
plt.imshow(img_segmentada_coche_azul)
plt.show()


#coche Rojo r
mask = cv.inRange(
    imgRGB, color_mean_r-color_std_r * n_sigma,  color_mean_r + color_std_r * n_sigma)
img_segmentada_coche_rojo = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche rojo")
plt.imshow(img_segmentada_coche_rojo)
plt.show()


#coche Amarillo am
mask = cv.inRange(
    imgRGB, color_mean_am-color_std_am * n_sigma,  color_mean_am + color_std_am * n_sigma)
img_segmentada_coche_amarillo = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche amarillo")
plt.imshow(img_segmentada_coche_amarillo)
plt.show()

#coche Marron m
mask = cv.inRange(
    imgRGB, color_mean_m-color_std_m * n_sigma,  color_mean_m + color_std_m * n_sigma)
img_segmentada_coche_marron = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche marron")
plt.imshow(img_segmentada_coche_marron)
plt.show()

#coche Verde Oscuro vo
mask = cv.inRange(
    imgRGB, color_mean_vo-color_std_vo * n_sigma,  color_mean_vo + color_std_vo * n_sigma)
img_segmentada_coche_verdeoscuro = cv.bitwise_and(imgRGB, imgRGB, mask=mask)
plt.figure("Segmentación de coche verde oscuro")
plt.imshow(img_segmentada_coche_verdeoscuro)
plt.show()

#img_suma = cv.add (img_segmentada_roja, img_segmentada_naranja, img_segmentada_azul)
img_suma = img_segmentada_coche_naranja + img_segmentada_coche_verde + img_segmentada_coche_turqueza + img_segmentada_coche_blanco + img_segmentada_coche_azul + img_segmentada_coche_rojo + img_segmentada_coche_amarillo + img_segmentada_coche_marron
plt.figure("Suma de las segmentaciones")
plt.imshow(img_suma)
plt.show
```
