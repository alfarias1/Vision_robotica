Generacion de metrica para todos los Frame del video

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Funcion para calcular la metrica FM
def calcular_fm(imagen):
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    f = np.fft.fft2(gris)     
    fshift = np.fft.fftshift(f)
    magnitud = np.abs(fshift)

    # Valor maximo y umbral
    M = np.max(magnitud)
    thres = M / 1000.0
    #se cuentan los valores que superan el umbral
    num_alto = np.sum(magnitud > thres)
    #Normalizamos por el total de pixeles
    fm = num_alto / magnitud.size
    return fm


video = cv2.VideoCapture('focus_video.mov')
fms = []
frames = []
#Para cada frame hago el calculo de la metrica
while True:
    ret, frame = video.read()
    if not ret:
        break
    fm = calcular_fm(frame)
    fms.append(fm)
    frames.append(frame)

video.release()

#busco el frame con maximo enfoque
max_index = np.argmax(fms)
print(f"Frame con maximo enfoque es: {max_index}, FM = {fms[max_index]:.5f}")

# Grafico
plt.figure(figsize=(12, 5))
plt.plot(fms, label='FM por frame')
plt.axvline(x=max_index, color='r', linestyle='--', label='Maximo enfoque')
plt.xlabel('Numero de frame')
plt.ylabel('Metrica FM')
plt.title('Evolucion de la nitidez en el video')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("evolucion_enfoque_frema_x_frame.png")
plt.show()
```

Medicion sobre una ROI ubicada en el centro del frame, area del ROI entre 5 o 10 % del area total del frame y generacion de video con la ROI marcada y color segun foco

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calcular_fm(imagen):
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    f = np.fft.fft2(gris)
    fshift = np.fft.fftshift(f)
    magnitud = np.abs(fshift)
    M = np.max(magnitud)
    thres = M / 1000.0
    num_alto = np.sum(magnitud > thres)
    fm = num_alto / magnitud.size
    return fm

def extraer_roi_central(frame, porcentaje=0.10):
    h, w = frame.shape[:2]
    factor = np.sqrt(porcentaje)
    roi_h = int(h * factor)
    roi_w = int(w * factor)
    start_y = h // 2 - roi_h // 2
    start_x = w // 2 - roi_w // 2
    roi = frame[start_y:start_y + roi_h, start_x:start_x + roi_w]
    return roi


video = cv2.VideoCapture('focus_video.mov')
fms = []
frames = []

while True:
    ret, frame = video.read()
    if not ret:
        break
    roi = extraer_roi_central(frame, porcentaje=0.10)
    fm = calcular_fm(roi)
    fms.append(fm)
    frames.append(frame)

video.release()

#Frame con maximo enfoque en la ROI
max_index = np.argmax(fms)
print(f"Frame con maximo enfoque (ROI): {max_index}, FM = {fms[max_index]:.5f}")

# Grafico
plt.figure(figsize=(12, 5))
plt.plot(fms, label='FM en ROI central')
plt.axvline(x=max_index, color='r', linestyle='--', label='Maximo enfoque')
plt.xlabel('Numero de frame')
plt.ylabel('Metrica de enfoque (FM en ROI)')
plt.title('Evolucion de la nitidez en ROI central')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("evolucion_enfoque_roi.png")
plt.show()


#Creacion del video con la ROI marcada

output = cv2.VideoWriter(
    'output_roi_focus.avi',
    cv2.VideoWriter_fourcc(*'XVID'),30,(frames[0].shape[1], frames[0].shape[0])
)

for i, frame in enumerate(frames):
    # Extraer ROI central del frame
    h, w = frame.shape[:2]
    factor = np.sqrt(0.10)
    roi_h = int(h * factor)
    roi_w = int(w * factor)
    start_y = h // 2 - roi_h // 2
    start_x = w // 2 - roi_w // 2

    # Color verde si esta en foco sino rojo 
    color = (0, 255, 0) if i == max_index else (0, 0, 255) 

    # Dibuja el rectangulo sobre el frame
    cv2.rectangle(
        frame,
        (start_x, start_y),
        (start_x + roi_w, start_y + roi_h),
        color,
        thickness=2
    )

    output.write(frame)

output.release()
print("Video generado: output_roi_focus.avi")
```

Medicion sobre una matriz de enfoque compuesta por un arreglo MxN elementos rectangualres equiespaciados. N y M son valores arbitrarios, probar con valores 5X5 ,7x5, 10X10 y Video de la matriz con foco

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calcular_fm(imagen):
    gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    f = np.fft.fft2(gris)
    fshift = np.fft.fftshift(f)
    magnitud = np.abs(fshift)
    M = np.max(magnitud)
    thres = M / 1000.0
    num_alto = np.sum(magnitud > thres)
    fm = num_alto / magnitud.size
    return fm

#Funcion para generar la grilla
def dividir_en_grilla(frame, N, M):
    h, w = frame.shape[:2]
    dh, dw = h // N, w // M
    celdas = []
    coords = []
    for i in range(N):
        for j in range(M):
            y1, y2 = i * dh, (i + 1) * dh
            x1, x2 = j * dw, (j + 1) * dw
            celdas.append(frame[y1:y2, x1:x2])
            coords.append(((x1, y1), (x2, y2)))
    return celdas, coords



# Parametros de grilla
N, M = 10, 10 

video = cv2.VideoCapture('focus_video.mov')
fms = []
frames = []

while True:
    ret, frame = video.read()
    if not ret:
        break
    celdas, _ = dividir_en_grilla(frame, N, M)
    fms_frame = [calcular_fm(celda) for celda in celdas]
    fms.append(np.mean(fms_frame))
    frames.append(frame)

video.release()

# Deteccion de maximo de enfoque
max_index = np.argmax(fms)
print(f"Frame con maximo enfoque en grilla {N}x{M}: {max_index}, FM promedio = {fms[max_index]:.5f}")

# Grafica
plt.figure(figsize=(12, 5))
plt.plot(fms, label=f'FM promedio - grilla {N}x{M}')
plt.axvline(x=max_index, color='r', linestyle='--', label='Maximo enfoque')
plt.xlabel('Frame')
plt.ylabel('FM promedio')
plt.title(f'Evolucion de nitidez con grilla {N}x{M}')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig(f"evolucion_enfoque_grilla_{N}x{M}.png")
plt.show()




# Creacion de video con grilla con color segun foco
output = cv2.VideoWriter(
    f'output_grilla_{N}x{M}.avi',
    cv2.VideoWriter_fourcc(*'XVID'),
    30,
    (frames[0].shape[1], frames[0].shape[0])
)

for i, frame in enumerate(frames):
    _, coords = dividir_en_grilla(frame, N, M)
    color = (0, 255, 0) if i == max_index else (0, 0, 255)
    for (x1, y1), (x2, y2) in coords:
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, thickness=1)
    output.write(frame)

output.release()
print(f"Video generado: output_grilla_{N}x{M}.avi")
```
